## Calculate the minimum distance between ER and HOPS

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tkinter import filedialog as fd
```

```python
# Function to load and process multiple CSV files
def load_and_process_files():
    file_paths = fd.askopenfilenames(filetypes=[("CSV files", "*.csv")])
    all_data = []
    for i, file_path in enumerate(file_paths):
        df = pd.read_csv(file_path)
        df['min_distance_to_ER'] = df['min_distance_to_ER'] * 117  # Convert to micrometers
        df['unique_condensate_id'] = f'replicate_{i}_' + df['trackID'].astype(str)
        all_data.append(df)
    return pd.concat(all_data, ignore_index=True)
```

```python
# Load and process the data
data = load_and_process_files()
# Count unique condensates
unique_condensates = data['unique_condensate_id'].nunique()
```

```python
# Create the plot
plt.figure(figsize=(6, 4))

ax = sns.histplot(
    data=data,
    x="min_distance_to_ER",
    bins=30,
    stat="probability",
    common_norm=False,
    binrange=(0, 200),  # Adjust range as needed
    lw=2,
    element="step",
    legend=False,
)

# plt.axvline(250, ls="--", c="gray", lw=2)
# plt.text(
#     275,
#     0.078,
#     r"N$_{condensates}$=" + f"{unique_condensates:,}",
#     fontdict={"fontsize": 16},
# )

plt.xlim(0, 200)  # Adjust as needed
plt.ylim(0, None)
plt.xlabel(r"Minimum Distance to ER (nm)", fontsize=18)
plt.ylabel("Probability", fontsize=18)
plt.gca().spines[:].set_linewidth(1)
plt.gca().tick_params(
    axis="both",
    which="major",
    labelsize=18,
    direction="in",
    bottom=True,
    left=True,
    length=5,
    width=1,
)
plt.gca().set_axisbelow(False)
plt.gca().tick_params(axis="both", which="major", labelsize=18)
plt.savefig("minimum_distance_ER_HOPS.png", format="png", bbox_inches="tight", dpi=600)
plt.show()
```

## Calculate the minimum distance between Lys and HOPS

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tkinter import filedialog as fd
```

```python
# Function to load and process multiple CSV files
def load_and_process_files():
    file_paths = fd.askopenfilenames(filetypes=[("CSV files", "*.csv")])
    all_data = []
    for i, file_path in enumerate(file_paths):
        df = pd.read_csv(file_path)
        df['min_distance_to_Lys'] = df['min_distance_to_Lys'] * 117  # Convert to micrometers
        df['unique_condensate_id'] = f'replicate_{i}_' + df['trackID'].astype(str)
        all_data.append(df)
    return pd.concat(all_data, ignore_index=True)
```

```python
# Load and process the data
data = load_and_process_files()
# Count unique condensates
unique_condensates = data['unique_condensate_id'].nunique()
```

```python
# Create the plot
plt.figure(figsize=(6, 4))

ax = sns.histplot(
    data=data,
    x="min_distance_to_Lys",
    bins=30,
    stat="probability",
    common_norm=False,
    binrange=(0, 500),  # Adjust range as needed
    lw=2,
    element="step",
    legend=False,
)

# plt.axvline(250, ls="--", c="gray", lw=2)
plt.text(
    275,
    0.078,
    r"N$_{condensates}$=" + f"{unique_condensates:,}",
    fontdict={"fontsize": 16},
)

plt.xlim(0, 500)  # Adjust as needed
plt.ylim(0, 0.09)
plt.xlabel(r"Minimum Distance to ER (nm)", fontsize=18)
plt.ylabel("Probability", fontsize=18)
plt.gca().spines[:].set_linewidth(1)
plt.gca().tick_params(
    axis="both",
    which="major",
    labelsize=18,
    direction="in",
    bottom=True,
    left=True,
    length=5,
    width=1,
)
plt.gca().set_axisbelow(False)
plt.gca().tick_params(axis="both", which="major", labelsize=18)
plt.savefig("minimum_distance_ER_HOPS.png", format="png", bbox_inches="tight", dpi=600)
plt.show()
```
