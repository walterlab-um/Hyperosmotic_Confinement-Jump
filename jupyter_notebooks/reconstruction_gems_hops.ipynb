```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data (replace paths with your actual files)
hops_path = r"Z:\Bisal_Halder_turbo\PROCESSED_DATA\Impact_of_cytoskeleton_on_HOPS_condensates\GEM tracking\GEM_HOPS Cotracking\Reconstruction image\linevalues_HOPS_MIP.csv"
gem_path = r"Z:\Bisal_Halder_turbo\PROCESSED_DATA\Impact_of_cytoskeleton_on_HOPS_condensates\GEM tracking\GEM_HOPS Cotracking\Reconstruction image\linevalues_GEM_track_occurance.csv"

hops_df = pd.read_csv(hops_path)
gem_df = pd.read_csv(gem_path)

# Create figure and primary axis
fig, ax1 = plt.subplots(figsize=(6, 5))

# Plot HOPS (RED, left Y-axis)
ax1.plot(
    hops_df["Distance_(microns)"],
    hops_df["Gray_Value"],
    color="#FF0000",  # Pure red
    lw=2,
    label="HOPS"
)
ax1.set_xlabel('Distance (m)', fontsize=18)
ax1.set_ylabel('DCP1A Intensity (A.U.)', fontsize=18, color="#FF0000")
ax1.tick_params(axis='y', labelcolor="#FF0000")

# Plot GEM (GREEN, right Y-axis)
ax2 = ax1.twinx()
ax2.plot(
    gem_df["Distance_(microns)"],
    gem_df["Gray_Value"],
    color="#000000",  # Pure green
    lw=1.5,
    label="GEM"
)
ax2.set_ylabel('GEM Trajectories Density (#/pxl.)', fontsize=18, color="#000000")
ax2.tick_params(axis='y', labelcolor="#000000")

# Aesthetics (matching your style)
plt.xlim(hops_df["Distance_(microns)"].min(), hops_df["Distance_(microns)"].max())
for ax in [ax1, ax2]:
    ax.spines[:].set_linewidth(1)
    ax.tick_params(
        axis="both",
        which="major",
        labelsize=18,
        direction="in",
        bottom=True,
        left=True,
        length=5,
        width=1,
    )

# Legend (place at upper right)
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
# ax1.legend(lines1 + lines2, labels1 + labels2, fontsize=16, frameon=False, loc='upper right')

plt.tight_layout()
plt.savefig("HOPS_red_GEM_green.png", format="png", bbox_inches="tight", dpi=600)
plt.show()
```
